```python
import pandas as pd
import glob
from joblib import Parallel, delayed
```

```python
files = glob.glob('base_problems_runs/*')
```

```python
dfs = Parallel(n_jobs=-1)(delayed(pd.read_csv)(f) for f in files)
df = pd.concat(dfs)
```

```python
df
```

```python
rdf = df[['GA', 'PSO', 'DE', 'CMAES', 'ES']].rank(axis=1)
rdf
```

```python
fdf = pd.concat([df, rdf], axis=1)
#fdf
```

```python
fdf.groupby(['problem_class']).mean().reset_index()
```

```python
d = fdf.groupby(['problem_class', 'problem_instance']).mean().reset_index()
d
```

```python
outdf = pd.concat([df[['problem_class']], rdf], axis=1).groupby(['problem_class']).mean().reset_index()
print(outdf.to_latex(index = False, multirow = True))
```
