```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)
y5 = np.sqrt(x)

# Create a figure with 5 subplots in one row
fig, axs = plt.subplots(1, 5, figsize=(9, 1.7))


mix = CombinedProblem(problem1, problem2, 1.0)
axs[0].imshow(generate_2d_plot_matrix(mix), cmap='viridis', extent=[-5, 5, -5, 5], origin='lower')
axs[0].scatter([mix.pareto_set()[0][0]], [mix.pareto_set()[0][1]], marker='x', color='red')

mix = CombinedProblem(problem1, problem2, 0.75)
axs[1].imshow(generate_2d_plot_matrix(mix), cmap='viridis', extent=[-5, 5, -5, 5], origin='lower')
axs[1].scatter([mix.pareto_set()[0][0]], [mix.pareto_set()[0][1]], marker='x', color='red')

mix = CombinedProblem(problem1, problem2, 0.5)
axs[2].imshow(generate_2d_plot_matrix(mix), cmap='viridis', extent=[-5, 5, -5, 5], origin='lower')
axs[2].scatter([mix.pareto_set()[0][0]], [mix.pareto_set()[0][1]], marker='x', color='red')

mix = CombinedProblem(problem1, problem2, 0.25)
axs[3].imshow(generate_2d_plot_matrix(mix), cmap='viridis', extent=[-5, 5, -5, 5], origin='lower')
axs[3].scatter([mix.pareto_set()[0][0]], [mix.pareto_set()[0][1]], marker='x', color='red')

mix = CombinedProblem(problem1, problem2, 0.0)
axs[4].imshow(generate_2d_plot_matrix(mix), cmap='viridis', extent=[-5, 5, -5, 5], origin='lower')
axs[4].scatter([mix.pareto_set()[0][0]], [mix.pareto_set()[0][1]], marker='x', color='red')

# Add titles to each subplot
axs[0].set_title('0.0')
axs[1].set_title('0.25')
axs[2].set_title('0.5')
axs[3].set_title('0.75')
axs[4].set_title('1.0')


for ax in axs:
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])
#axs[1].set_title('cos(x)')
#axs[2].set_title('tan(x)')
#axs[3].set_title('exp(-x)')
#axs[4].set_title('sqrt(x)')

# Add a title for the whole figure
#fig.suptitle('Five Subplots in One Row')

# Show the plot
fig.subplots_adjust(wspace=0)
plt.savefig(f'figures/function_mix.pdf', bbox_inches='tight', pad_inches=0)
```
