```python
def generate_candidate_solutions(limx, limy):
    samples_x = np.linspace(limx[0], limx[1], 200)
    samples_y = np.linspace(limy[0], limy[1], 200)
    X = []
    for ii, i in enumerate(samples_x):
        for ij, j in enumerate(samples_y):
              X.append([i, j])
    return np.array(X)

def plot(problem):
    samples_x = np.linspace(-5, 5, 200)
    samples_y = np.linspace(-5, 5, 200)
    
    X = []
    P = []
    for i, x in enumerate(samples_x):
        for j, y in enumerate(samples_y):
            P.append((i, j))
            X.append((x, y))
    X = np.array(X)
    P = np.array(P) 
    
    y = problem.evaluate(X)
    
    M = np.zeros((len(samples_x), len(samples_y)))
    for pi, yi in zip(P, y):
        M[pi[0], pi[1]] = yi
    return M
```

```python
problem1 = COCO(1, 1, 2)
problem2 = COCO(21, 1, 2)
mix2 = ProblemMix(problem1, problem2, mix=0.8)
mix4 = ProblemMix(problem1, problem2, mix=0.6)
mix6 = ProblemMix(problem1, problem2, mix=0.4)
mix8 = ProblemMix(problem1, problem2, mix=0.2)

fig, axes = plt.subplots(2, 3, figsize=(10, 7), sharey=False)

axes[0, 0].imshow(plot(problem1))
axes[0, 1].imshow(plot(mix2))
axes[0, 2].imshow(plot(mix4))
axes[1, 0].imshow(plot(mix6))
axes[1, 1].imshow(plot(mix8))
axes[1, 2].imshow(plot(problem2))
plt.savefig(f'figures/mixed_problems.pdf', bbox_inches='tight', pad_inches=0)
```
